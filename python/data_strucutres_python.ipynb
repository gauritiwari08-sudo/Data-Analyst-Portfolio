List and all features or functions

```python
#creating a list
my_list = [25, "Hello", 3.14, "Python", 100, 7.89, "AI", 42, "Data", 2.718]
print(my_list)

#indexing
positive_indexing = [my_list[0], my_list[3], my_list[6]]
print(positive_indexing)
negative_indexing = [my_list[-1], my_list[-4], my_list[-7]]
print(negative_indexing)

#slicing
sublist = my_list[2:7]
print(sublist)

#adding element using appned()
my_list.append("New Element")
print(my_list)

#removing particular element from list remove()
my_list.remove(100)
print(my_list)

#removing element at particular index from list remove()
my_list.pop(4)
print(my_list)

#adding another list or multiple elements to list
my_list.extend([100,200,300])
print(my_list)


```

Tuple and all its functions

```python
#creating tuple
my_tuple = (10, "Python", 3.14, "AI", 42, "Data", 7.89, "Machine Learning")
print(my_tuple)

#accessing elements
positive_indexing = my_tuple[1]  # "Python"
print(positive_indexing)
negative_indexing = my_tuple[-2]  # 7.89
print(negative_indexing)

#slicing
subtuple = my_tuple[2:6]
print(subtuple)

#Tuple packing : assigning multiple values to single tuple
person = ("John", 25, "Developer")
print(person) # Output: ('John', 25, 'Developer')

#tuple unpacking
name,age,designation = person
print(name) # Output: John
print(age) # Output: 25
print(designation) # Output: Developer

#convert tuple to list
my_list=list(my_tuple)
print(my_list)
my_list.append("New Element")
print(my_list)


#convert list to tuple
modified_my_tuple=tuple(my_list)
print(modified_my_tuple)

```
