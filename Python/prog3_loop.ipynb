Practice program for loops

```python
#Print numbers from 1 to 10
for i in range(1,11):
  print(i)
```

```python
#Print numbers from 10 to 1
for i in range(10,0,-1):
  print(i)
```

```python
#Print even numbers from 1 to 100
for i in range(2,101,2):
  print(i)
```

```python
#Print the multiplication table of any number
num=int( input("Enter any no to display its table"))
for i in range(1,11):
  print(i*num)
```

```python
#Find the sum of the first 100 natural numbers
sum=0
for i in range(1,101):
  sum=sum+i
print(sum)
```

```python
#Find the factorial of a number
n=int(input("Enter any no to find its factorial"))
fact=1
for i in range(1,n+1):
  fact=fact*i
print(fact)
```

```python
#Reverse a number
n=int(input("Enter any no to reverse"))
rev=0
while n>0:
  rem=n%10
  rev=rev*10+rem
  n=n//10
print(rev)
```

```python
#Count the digits in a number
n=int(input("Enter any no to count its digits"))
count=0
while n>0:
  count=count+1
  n=n//10
print(count)
```

```python
#Check whether a number is a palindrome
num=int(input("Enter any no to check whether it is palindrome or not"))
rev=0
n=num
while n>0:
  rem=n%10
  rev=rev*10+rem
  n=n//10
#print(rev,num)
if(rev==num):
  print("palindrome")
else:
  print("NOt PAlindrome")
```
