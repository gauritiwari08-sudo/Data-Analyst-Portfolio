List and all features or functions

```python
#creating a list
my_list = [25, "Hello", 3.14, "Python", 100, 7.89, "AI", 42, "Data", 2.718]
print(my_list)

#indexing
positive_indexing = [my_list[0], my_list[3], my_list[6]]
print(positive_indexing)
negative_indexing = [my_list[-1], my_list[-4], my_list[-7]]
print(negative_indexing)

#slicing
sublist = my_list[2:7]
print(sublist)

#adding element using appned()
my_list.append("New Element")
print(my_list)

#removing particular element from list remove()
my_list.remove(100)
print(my_list)

#removing element at particular index from list remove()
my_list.pop(4)
print(my_list)

#adding another list or multiple elements to list
my_list.extend([100,200,300])
print(my_list)


```

Tuple and all its functions

```python
#creating tuple
my_tuple = (10, "Python", 3.14, "AI", 42, "Data", 7.89, "Machine Learning")
print(my_tuple)

#accessing elements
positive_indexing = my_tuple[1]  # "Python"
print(positive_indexing)
negative_indexing = my_tuple[-2]  # 7.89
print(negative_indexing)

#slicing
subtuple = my_tuple[2:6]
print(subtuple)

#Tuple packing : assigning multiple values to single tuple
person = ("John", 25, "Developer")
print(person) # Output: ('John', 25, 'Developer')

#tuple unpacking
name,age,designation = person
print(name) # Output: John
print(age) # Output: 25
print(designation) # Output: Developer

#convert tuple to list
my_list=list(my_tuple)
print(my_list)
my_list.append("New Element")
print(my_list)


#convert list to tuple
modified_my_tuple=tuple(my_list)
print(modified_my_tuple)

```

Dictionary

```python
#creating a dictionary
student_dict = {
    "name": "John Doe",
    "age": 21,
    "course": "Computer Science",
    "marks": 85,
    "city": "New York"
}

print(student_dict)

#Accesing element using keys
student_name = student_dict["name"]
student_course = student_dict["course"]

print(student_name)
print(student_course)

#modifing a value
student_dict["age"] = 22
print(student_dict)

#adding a new key value pair
student_dict["grade"] = "A"
print(student_dict)

#accessing element using get method
print(student_dict.get("age"))
print(student_dict.get("address"),"Not Found")

#deleting a key value pair
del student_dict["grade"]
print(student_dict)

#using pop funciton
student_dict.pop("city")
print(student_dict)

#clear all dictionary
student_dict.clear()
print(student_dict)


```

```python
#nested dictionary
# Nested dictionary
company = {
 "CEO": {"name": "Alice", "age": 45},
 "Manager": {"name": "Bob", "age": 38},
 "Intern": {"name": "Charlie", "age": 22}
}

print(company)
```

SET

```python
# Creating a set with unique elements
my_set = {10, "Python", 3.14, "AI", 42, "Data", 7.89}
print(my_set)

my_set.add("Machine Learning")  # Adding a new element
print(my_set)
my_set.remove(42)  # Removing an existing element
print(my_set)

my_set.add("Python")  # "Python" already exists, so no change
print(my_set)

another_set = {3.14, "AI", "Deep Learning", 99, "Data"}

union_set = my_set.union(another_set)  # Combines elements from both sets
intersection_set = my_set.intersection(another_set)  # Common elements
difference_set = my_set.difference(another_set)  # Elements in my_set but not in another_set
```

```python
my_list_with_duplicates = [1, 2, 2, 3, 4, 4, 5, 1, 6]
print("Original list:", my_list_with_duplicates)

# Convert the list to a set to remove duplicates
# Then convert it back to a list using __builtins__.list to avoid conflicts
unique_elements_list = __builtins__.list(set(my_list_with_duplicates))
print("List after removing duplicates:", unique_elements_list)
```
